# Removing Multicolinearity using VIF

**A rule of thumb commonly used in practice is if a VIF is > 5, you have high multicollinearity.**

```
def calculate_vif_(X, thresh=5.0):
    variables = list(range(X.shape[1]))
    dropped = True
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(X.iloc[:, variables].values, ix)
               for ix in range(X.iloc[:, variables].shape[1])]
        max_vif_val = max(vif)
        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print('Dropping \'' + X.iloc[:, variables].columns[maxloc] +
                  '\' with VIF value : ' + str(max_vif_val))
            del variables[maxloc]
            dropped = True

    print('Remaining variables:')
    print(X.columns[variables])
    return X.iloc[:, variables]
```

# Linear Regression using OLS (StatsModel)

```
X = sm.add_constant(X_train)
model = sm.OLS(y_train, X)
```

```
results = model.fit()
results.params
```

```
print(results.summary())
```

# Observation based on the generated model

*   We start the exercise by using the statsmodel based OLS model to fit X & Y because of the results provided by the model that explain the significance of each predictor variable.
*   The F-test of overall significance indicates whether your linear regression model provides a better fit to the data than a model that contains no independent variables. So, in our case based on the p-value of F-test or Prob (F-statistic) we conclude that the model is a good fit model.
*   R-squared value of 0.69 signifies it a descent fit model.
*   The P>|t| or the p-value of the feature signifies how significant a feature is. 
*   In Linear Regression, the Null Hypothesis for a feature & target variable is that there is no relationship between target & the feature. 
* Alternate Hypothesis states that there is no relationship between target and feature.
* A p-value of less than 0.05 signifies that the features **Feature 1, 4, 13** are signifiant predictors and are related with the target variable.
* However, **Feature 15** has a p-value of 0.064 which makes it insignificant for the given target variable.

# Linear Regression using Ridge Regression

```
# Ridge regression is a technique to reduce model complexity & 
# prevent over-fitting which may result from simple linear regression.
```

```
ridge = Ridge()
params = {"alpha" : [1e-15, 1e-10, 1e-8, 1e-4, 1e-3, 1e-2, 1, 5, 10, 20]}
```

```
ridge_regressor = GridSearchCV(ridge, params, scoring="neg_mean_squared_error", cv=5)
```

```
ridge_regressor.fit(X_train, y_train)
```

```
ridge_regressor.best_score_
```

```
ridge_regressor.best_estimator_
```

```
ridge_best = ridge_regressor.best_estimator_
```

```
# fit the entire training data on the best estimator
ridge_best.fit(X_train, y_train)
```

```
coef = pd.Series(ridge_best.coef_,list(X_train.columns))
coef.plot(kind='bar', title='Model Coefficients')
```

```
X_test_without_const = X_test.drop(["const"], axis=1)
```

```
y_ridge_pred = ridge_best.predict(X_test_without_const)
```

```
mean_squared_error(y_test, y_ridge_pred)
```

```
r2_score(y_test, y_ridge_pred)
```
